## 0. Setup e importaciones

Importamos las librerías necesarias y configuramos el estilo de los gráficos.

- **pandas / numpy**: manipulación de datos
- **matplotlib**: gráficos base
- **seaborn**: gráficos estadísticos (heatmaps, scatter)
- **warnings**: silenciamos avisos no críticos para mantener la salida limpia

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# --- Configuración visual ---
# Definimos el tamaño por defecto de los gráficos y el estilo
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['font.size'] = 11
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.titleweight'] = 'bold'
sns.set_style('whitegrid')   # fondo blanco con grilla gris

# Paleta de colores consistente para todo el notebook
# Usamos colores con significado: rojo = dengue (alarma), naranja = calor,
# azul = lluvia, verde = humedad
COLOR_DENGUE  = '#E74C3C'
COLOR_TEMP    = '#E67E22'
COLOR_LLUVIA  = '#3498DB'
COLOR_HUMEDAD = '#27AE60'

print('Setup completado correctamente')
```

---
## 1. Carga del dataset maestro

Cargamos el archivo `dataset_maestro.parquet` generado en el Sprint 2.
Este dataset contiene:
- **2,340 filas**: 15 comunas × 52 semanas × 3 años
- **13 columnas**: casos de dengue + variables climáticas + variables derivadas

Lo primero que hacemos es ver las primeras filas para confirmar que todo esté bien.

```python
# Cargamos el dataset maestro
# read_parquet es más rápido que read_csv y preserva los tipos de datos
df = pd.read_parquet('../data/processed/dataset_maestro.parquet')

# Información básica del dataset
print('=== Información del dataset maestro ===')
print(f'Filas:    {len(df):,}')
print(f'Columnas: {len(df.columns)}')
print(f'Años:     {sorted(df.year.unique())}')
print(f'Comunas:  {sorted(df.comuna_id.unique())}')
print(f'Semanas:  {df.epi_week.min()} a {df.epi_week.max()}')
print()
print('Primeras filas:')
df.head(10)
```

**¿Qué observamos?**
- Cada fila representa una combinación única de año + semana + comuna
- Las variables climáticas son iguales para todas las comunas en la misma semana (usamos una sola estación meteorológica para toda CABA)
- La columna `confirmed_cases` es nuestra variable objetivo — lo que el modelo va a predecir

---
## 2. Serie temporal de casos — CABA total

El primer gráfico muestra la evolución de los casos a lo largo del tiempo.
Para esto necesitamos **agregar** (sumar) los casos de las 15 comunas en cada semana.

**¿Por qué hacemos esto?**
Porque con 15 comunas × 156 semanas sería muy difícil ver el patrón global. Primero entendemos el comportamiento de CABA como un todo y después vemos las diferencias entre comunas.

```python
# Paso 1: Agregar casos sumando todas las comunas por semana
# groupby(['year', 'epi_week']) agrupa todas las filas del mismo año y semana
# .sum() suma los casos de todas las comunas
# .first() toma el primer valor climático (son iguales para todas las comunas)
df_total = (
    df.groupby(['year', 'epi_week'], as_index=False)
    .agg(
        casos_total=('confirmed_cases', 'sum'),
        temp_mean=('temp_mean', 'first'),
        precipitation=('precipitation', 'first'),
        humidity_mean=('humidity_mean', 'first'),
    )
)

# Paso 2: Crear etiqueta de tiempo para el eje X
# Formato: "2023-SE01", "2023-SE02", etc.
df_total['tiempo'] = (
    df_total['year'].astype(str) + '-SE' +
    df_total['epi_week'].astype(str).str.zfill(2)
)

# Paso 3: Ordenar cronológicamente y crear índice numérico para el eje X
df_total = df_total.sort_values(['year', 'epi_week']).reset_index(drop=True)
df_total['idx'] = range(len(df_total))

print(f'Total de semanas: {len(df_total)}')
print(f'Total de casos en el período: {df_total["casos_total"].sum():,}')
```

```python
# --- Gráfico de serie temporal ---
fig, ax = plt.subplots(figsize=(16, 5))

# fill_between: rellena el área bajo la curva con color semitransparente
# Ayuda a percibir visualmente la magnitud de los brotes
ax.fill_between(
    df_total['idx'],
    df_total['casos_total'],
    alpha=0.3, color=COLOR_DENGUE
)
# La línea encima del área rellena marca el valor exacto cada semana
ax.plot(
    df_total['idx'],
    df_total['casos_total'],
    color=COLOR_DENGUE, linewidth=1.5, label='Casos confirmados'
)

# Anotamos el pico máximo con una flecha
# idxmax() devuelve el índice de la fila con el valor más alto
idx_max = df_total['casos_total'].idxmax()
ax.annotate(
    f"Pico: {df_total.loc[idx_max, 'casos_total']:,} casos\n"
    f"{df_total.loc[idx_max, 'tiempo']}",
    xy=(idx_max, df_total.loc[idx_max, 'casos_total']),
    xytext=(idx_max - 8, df_total.loc[idx_max, 'casos_total'] * 0.85),
    arrowprops=dict(arrowstyle='->', color='black'),
    fontsize=10, fontweight='bold'
)

# Sombreamos la temporada alta (SE 1-17 = enero a abril)
# que es cuando históricamente hay más transmisión de dengue en CABA
for year_idx, year in enumerate(sorted(df_total['year'].unique())):
    mask_alta = (df_total['year'] == year) & (df_total['epi_week'] <= 17)
    if mask_alta.any():
        x_ini = df_total.loc[mask_alta, 'idx'].min()
        x_fin = df_total.loc[mask_alta, 'idx'].max()
        ax.axvspan(x_ini, x_fin, alpha=0.08, color='red',
                   label='Temporada alta (SE 1-17)' if year_idx == 0 else '')

# Eje X: mostramos solo el año (no todas las semanas, sería ilegible)
ticks_año = df_total[df_total['epi_week'] == 1]['idx'].tolist()
labels_año = df_total[df_total['epi_week'] == 1]['year'].astype(str).tolist()
ax.set_xticks(ticks_año)
ax.set_xticklabels(labels_año)

ax.set_xlabel('Año')
ax.set_ylabel('Casos confirmados de dengue')
ax.set_title('Casos confirmados de dengue — CABA total (2023-2025)')
ax.legend(loc='upper left')
# Formateamos los números del eje Y con separador de miles
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{int(x):,}'))

plt.tight_layout()
plt.savefig('../reports/figures/01_serie_temporal_caba.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Pico máximo: {df_total["casos_total"].max():,} casos en {df_total.loc[idx_max, "tiempo"]}')
```

**¿Qué observamos?**
- Los brotes ocurren claramente en verano (áreas sombreadas en rojo)
- El brote de 2024 fue significativamente más grande que el de 2023
- Entre brotes (invierno) los casos caen casi a cero
- Esto confirma la fuerte estacionalidad del dengue en CABA

## 3. Comparación entre años

Superponemos los tres años en el mismo gráfico usando la semana epidemiológica como eje X.

**¿Por qué superponerlos?**
Porque así podemos comparar directamente si el brote de 2024 empezó antes, fue más intenso o duró más que el de 2023. Es una forma de ver la estacionalidad.

```python
fig, ax = plt.subplots(figsize=(14, 5))

# Un color distinto por año para distinguirlos fácilmente
colores_años = {2023: '#3498DB', 2024: '#E74C3C', 2025: '#27AE60'}

for year in sorted(df_total['year'].unique()):
    datos_año = df_total[df_total['year'] == year].copy()
    total_año = datos_año['casos_total'].sum()
    ax.plot(
        datos_año['epi_week'],
        datos_año['casos_total'],
        color=colores_años.get(year, 'gray'),
        linewidth=2,
        marker='o', markersize=3,  # puntos pequeños en cada semana
        label=f'{year} ({total_año:,} casos totales)'
    )

# Sombreamos la temporada alta y marcamos su fin con línea punteada
ax.axvspan(1, 17, alpha=0.08, color='red', label='Temporada alta (SE 1-17)')
ax.axvline(x=17, color='red', linestyle='--', alpha=0.4, linewidth=1)

ax.set_xlabel('Semana epidemiológica')
ax.set_ylabel('Casos confirmados (CABA total)')
ax.set_title('Comparación de casos por semana epidemiológica — 2023 vs 2024 vs 2025')
ax.legend()
ax.set_xlim(1, 52)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{int(x):,}'))

plt.tight_layout()
plt.savefig('../reports/figures/02_comparacion_años.png', dpi=150, bbox_inches='tight')
plt.show()
```

**¿Qué observamos?**
- El pico de ambos años ocurre aproximadamente en la misma semana (SE 12-13 = marzo)
- El brote de 2024 fue mucho más intenso — el pico es casi el doble que el de 2023
- 2025 tiene muy pocos datos porque el dataset solo cubre parte del año
- La forma de la curva es similar en ambos años (campana con pico en marzo), lo que confirma la estacionalidad

## 4. Distribución de casos por comuna

Ahora analizamos la dimensión **espacial** — qué comunas tienen más casos.

Creamos dos gráficos:
1. **Barplot total**: casos acumulados por comuna en todo el período
2. **Barplot por año**: cómo cambia la distribución entre 2023 y 2024

```python
# Total de casos por comuna en todo el período
# sort_values ordena de mayor a menor para que sea más fácil identificar
# las comunas más afectadas
df_por_comuna = (
    df.groupby('comuna_id')['confirmed_cases']
    .sum()
    .reset_index()
    .sort_values('confirmed_cases', ascending=False)
)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# --- Gráfico 1: total por comuna ---
ax = axes[0]
bars = ax.bar(
    df_por_comuna['comuna_id'].astype(str),
    df_por_comuna['confirmed_cases'],
    color=COLOR_DENGUE, alpha=0.8, edgecolor='white'
)
# Resaltamos la barra con más casos en rojo más oscuro
bars[0].set_color('#8B0000')

ax.set_xlabel('Comuna')
ax.set_ylabel('Casos confirmados totales')
ax.set_title('Total de casos por comuna (2023-2025)')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{int(x):,}'))

# --- Gráfico 2: por año ---
# Pivot para tener columnas = años
ax2 = axes[1]
df_comuna_año = (
    df.groupby(['comuna_id', 'year'])['confirmed_cases']
    .sum()
    .reset_index()
)

# Agrupamos barras: para cada comuna mostramos 3 barras (2023, 2024, 2025)
# width controla el ancho de cada barra
# El desplazamiento (i * width) evita que se superpongan
width = 0.25
comunas = sorted(df['comuna_id'].unique())
x = np.arange(len(comunas))  # posiciones numéricas para cada comuna

for i, (year, color) in enumerate([(2023, '#3498DB'), (2024, '#E74C3C'), (2025, '#27AE60')]):
    datos = df_comuna_año[df_comuna_año['year'] == year].set_index('comuna_id')
    valores = [datos.loc[c, 'confirmed_cases'] if c in datos.index else 0 for c in comunas]
    ax2.bar(x + i * width, valores, width, label=str(year), color=color, alpha=0.8)

ax2.set_xlabel('Comuna')
ax2.set_ylabel('Casos confirmados')
ax2.set_title('Casos por comuna y año')
ax2.set_xticks(x + width)
ax2.set_xticklabels([str(c) for c in comunas])
ax2.legend(title='Año')
ax2.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{int(x):,}'))

plt.tight_layout()
plt.savefig('../reports/figures/03_casos_por_comuna.png', dpi=150, bbox_inches='tight')
plt.show()

print('Top 5 comunas con más casos:')
print(df_por_comuna.head().to_string(index=False))
```

## 5. Heatmap: casos por semana epidemiológica y comuna

El heatmap es una visualización bidimensional donde:
- **Eje Y**: semana epidemiológica (1 a 52)
- **Eje X**: comunas (1 a 15)
- **Color**: número de casos (más rojo = más casos)

Hacemos uno por año para ver si el patrón espacial cambia entre años.

**¿Qué es `pivot_table`?**
Transforma el DataFrame de formato largo (una fila por semana+comuna) a formato ancho (una fila por semana, una columna por comuna). Es necesario para poder graficar el heatmap.

```python
# Creamos un subplot con 3 heatmaps, uno por año
fig, axes = plt.subplots(1, 3, figsize=(20, 7))

for ax, year in zip(axes, [2023, 2024, 2025]):
    # Filtramos el año y creamos la tabla pivote
    df_año = df[df['year'] == year]
    pivot = df_año.pivot_table(
        index='epi_week',       # semanas en las filas
        columns='comuna_id',    # comunas en las columnas
        values='confirmed_cases',
        aggfunc='sum',
        fill_value=0            # semanas sin casos = 0 (no NaN)
    )

    # YlOrRd = escala de color Amarillo → Naranja → Rojo
    # Intuitivo: amarillo = pocos casos, rojo = muchos casos
    sns.heatmap(
        pivot,
        ax=ax,
        cmap='YlOrRd',
        cbar_kws={'label': 'Casos'},
        linewidths=0.1,
        linecolor='white'
    )

    # Sombreamos las semanas de temporada alta (SE 1-17)
    ax.axhspan(0, 17, alpha=0.15, color='red')

    ax.set_title(f'Año {year}', fontweight='bold')
    ax.set_xlabel('Comuna')
    ax.set_ylabel('Semana epidemiológica')

plt.suptitle(
    'Heatmap: casos de dengue por semana epidemiológica y comuna — CABA\n'
    '(área sombreada = temporada alta SE 1-17)',
    fontsize=13, fontweight='bold', y=1.02
)
plt.tight_layout()
plt.savefig('../reports/figures/04_heatmap_semana_comuna.png', dpi=150, bbox_inches='tight')
plt.show()
```

**¿Qué observamos?**
- En ambos años los casos se concentran en las primeras semanas del año (área sombreada)
- El heatmap de 2024 tiene colores más intensos (más rojos), confirmando el brote más grande
- Hay variabilidad entre comunas — algunas columnas son más oscuras que otras
- La distribución espacial no es uniforme: algunas comunas tienen picos más pronunciados

## 6. Estacionalidad — promedio de casos por semana del año

Calculamos el **promedio histórico** de casos para cada semana del año, promediando entre los 3 años disponibles.

**¿Para qué sirve esto?**
Para identificar en qué semanas el dengue es endémicamente más activo, independientemente de si fue un año epidémico o no. Este patrón le dará al modelo información sobre la temporada del año.

```python
# Paso 1: Para cada año, sumar casos de todas las comunas por semana
# Paso 2: Promediar entre años para obtener el patrón estacional típico
df_estacional = (
    df.groupby(['year', 'epi_week'])['confirmed_cases']
    .sum()          # suma comunas → total CABA por semana
    .reset_index()
    .groupby('epi_week')['confirmed_cases']
    .mean()         # promedio entre años
    .reset_index()
)

fig, ax = plt.subplots(figsize=(14, 5))

# Coloreamos las barras: rojo = temporada alta, gris = temporada baja
ax.bar(
    df_estacional['epi_week'],
    df_estacional['confirmed_cases'],
    color=[
        COLOR_DENGUE if w <= 17 or w >= 48
        else '#BDC3C7'   # gris para temporada baja
        for w in df_estacional['epi_week']
    ],
    alpha=0.85,
    edgecolor='white'
)

# Sombreamos las zonas de temporada alta
ax.axvspan(0.5, 17.5, alpha=0.08, color='red', label='Temporada alta (SE 1-17)')
ax.axvspan(47.5, 52.5, alpha=0.08, color='red')

# Línea vertical en la semana de pico histórico
semana_pico = df_estacional.loc[
    df_estacional['confirmed_cases'].idxmax(), 'epi_week'
]
ax.axvline(
    x=semana_pico, color='darkred', linestyle='--',
    linewidth=1.5, label=f'Pico histórico (SE {int(semana_pico)})'
)

ax.set_xlabel('Semana epidemiológica')
ax.set_ylabel('Casos promedio (CABA total)')
ax.set_title('Estacionalidad del dengue en CABA — promedio histórico por semana')
ax.set_xlim(0.5, 52.5)
ax.legend()

# Eje X superior con nombres de meses (referencia aproximada)
meses = {
    1: 'Ene', 5: 'Feb', 9: 'Mar', 13: 'Abr', 18: 'May',
    22: 'Jun', 27: 'Jul', 31: 'Ago', 36: 'Sep',
    40: 'Oct', 45: 'Nov', 49: 'Dic'
}
ax2 = ax.twiny()   # segundo eje X en la parte superior
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks(list(meses.keys()))
ax2.set_xticklabels(list(meses.values()), fontsize=9)
ax2.tick_params(top=False)

plt.tight_layout()
plt.savefig('../reports/figures/05_estacionalidad.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Semana de pico histórico: SE {int(semana_pico)}')
print(f'Casos promedio en el pico: {df_estacional["confirmed_cases"].max():.0f}')
print(f'Casos promedio en invierno (SE 25): '
      f'{df_estacional[df_estacional["epi_week"]==25]["confirmed_cases"].values[0]:.0f}')
```

## 7. Correlaciones clima-dengue con rezagos

Este es uno de los análisis más importantes del EDA porque responde la pregunta central:
**¿Cuántas semanas después de un cambio climático se ve el efecto en los casos de dengue?**

**¿Qué es la correlación de Pearson?**
Es una medida de qué tan relacionadas están dos variables:
- r = 1: perfectamente correlacionadas positivamente (cuando sube una, sube la otra)
- r = -1: perfectamente correlacionadas negativamente
- r = 0: no hay correlación lineal
- |r| > 0.3: correlación moderada, relevante para el modelo

**¿Qué es un rezago (lag)?**
Calculamos la correlación entre los casos de la semana actual y el clima de hace 0, 1, 2, 3 y 4 semanas. Si el lag 2 tiene la correlación más alta, significa que el clima de hace 2 semanas es el más predictivo.

```python
# Trabajamos con datos CABA total (suma de comunas)
df_corr = df_total[['casos_total', 'temp_mean', 'precipitation', 'humidity_mean']].copy()

# Calculamos la correlación para cada variable climática y cada lag
variables_clima = ['temp_mean', 'precipitation', 'humidity_mean']
lags = range(0, 5)  # de lag 0 (misma semana) a lag 4 (4 semanas antes)

resultados = []
for var in variables_clima:
    for lag in lags:
        # .shift(lag) desplaza la columna lag semanas hacia abajo
        # Esto alinea el clima de hace 'lag' semanas con los casos actuales
        serie_lag = df_corr[var].shift(lag)
        corr = df_corr['casos_total'].corr(serie_lag)
        resultados.append({'variable': var, 'lag': lag, 'correlacion': corr})

df_corr_result = pd.DataFrame(resultados)

# --- Gráfico ---
fig, ax = plt.subplots(figsize=(12, 5))

colores_vars = {
    'temp_mean': COLOR_TEMP,
    'precipitation': COLOR_LLUVIA,
    'humidity_mean': COLOR_HUMEDAD
}
labels_vars = {
    'temp_mean': 'Temperatura media',
    'precipitation': 'Precipitación',
    'humidity_mean': 'Humedad relativa'
}

for var in variables_clima:
    datos_var = df_corr_result[df_corr_result['variable'] == var]
    ax.plot(
        datos_var['lag'],
        datos_var['correlacion'],
        marker='o', markersize=8,
        linewidth=2,
        color=colores_vars[var],
        label=labels_vars[var]
    )
    # Anotamos el valor de correlación máxima para cada variable
    idx_max = datos_var['correlacion'].abs().idxmax()
    ax.annotate(
        f"{datos_var.loc[idx_max, 'correlacion']:.2f}",
        xy=(datos_var.loc[idx_max, 'lag'], datos_var.loc[idx_max, 'correlacion']),
        textcoords='offset points', xytext=(5, 5),
        fontsize=9, color=colores_vars[var]
    )

# Líneas de referencia
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.8, alpha=0.5)
# ±0.3 es el umbral convencional para correlación moderada
ax.axhline(y=0.3, color='gray', linestyle='--', linewidth=0.8, alpha=0.5,
           label='Umbral correlación moderada (±0.3)')
ax.axhline(y=-0.3, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)

ax.set_xlabel('Rezago (semanas)')
ax.set_ylabel('Correlación de Pearson (r)')
ax.set_title('Correlación entre variables climáticas y casos de dengue\npor rezago temporal')
ax.set_xticks(list(lags))
ax.set_xticklabels([f'Lag {l}w' for l in lags])
ax.legend()
ax.set_ylim(-1, 1)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../reports/figures/06_correlaciones_clima_dengue.png', dpi=150, bbox_inches='tight')
plt.show()

print('Correlaciones máximas por variable climática:')
print('-' * 55)
for var in variables_clima:
    datos_var = df_corr_result[df_corr_result['variable'] == var]
    idx_max = datos_var['correlacion'].abs().idxmax()
    print(
        f'  {labels_vars[var]:<22}: r={datos_var.loc[idx_max,"correlacion"]:+.3f} '
        f'(lag {int(datos_var.loc[idx_max,"lag"])}w)'
    )
```

**¿Cómo interpretamos este gráfico?**
- Si la correlación máxima de temperatura está en **lag 2**, significa que la temperatura de hace 2 semanas predice mejor los casos actuales
- Esto tiene sentido biológico: el mosquito tarda ~2 semanas en completar su ciclo de vida y transmitir el virus
- Estos lags óptimos son los que usaremos como features en el Sprint 4

## 8. Scatter: relación directa clima vs casos

Graficamos cada variable climática contra los casos en un scatter plot.
Cada punto es una semana. Añadimos una línea de tendencia (regresión lineal) para ver la dirección de la relación.

**¿Qué esperamos ver?**
- Temperatura: relación positiva (más calor → más casos)
- Precipitación: relación positiva (más lluvia → más criaderos → más casos)
- Humedad: relación positiva (más humedad → mayor supervivencia del mosquito)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

pares = [
    ('temp_mean', 'Temperatura media (°C)', COLOR_TEMP),
    ('precipitation', 'Precipitación (mm/semana)', COLOR_LLUVIA),
    ('humidity_mean', 'Humedad relativa (%)', COLOR_HUMEDAD),
]

for ax, (var, label, color) in zip(axes, pares):

    # Graficamos puntos con forma y color distintos por año
    # Esto permite ver si la relación es consistente entre años
    for year, marker in [(2023, 'o'), (2024, 's'), (2025, '^')]:
        mask = df_total['year'] == year
        ax.scatter(
            df_total.loc[mask, var],
            df_total.loc[mask, 'casos_total'],
            alpha=0.6, s=40, marker=marker,
            label=str(year)
        )

    # Línea de tendencia usando regresión lineal (polyfit de grado 1)
    # np.polyfit ajusta una recta y = ax + b a los datos
    # np.poly1d convierte los coeficientes en una función evaluable
    x = df_total[var].values
    y = df_total['casos_total'].values
    mask_valid = ~np.isnan(x) & ~np.isnan(y)
    z = np.polyfit(x[mask_valid], y[mask_valid], 1)
    p = np.poly1d(z)
    x_line = np.linspace(x[mask_valid].min(), x[mask_valid].max(), 100)
    ax.plot(
        x_line, p(x_line),
        color='black', linewidth=1.5, linestyle='--',
        alpha=0.7, label='Tendencia'
    )

    # Calculamos y mostramos la correlación en el título
    corr = pd.Series(x[mask_valid]).corr(pd.Series(y[mask_valid]))
    ax.set_xlabel(label)
    ax.set_ylabel('Casos CABA total')
    ax.set_title(f'r = {corr:.3f}')
    ax.legend(fontsize=8)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{int(x):,}'))

plt.suptitle(
    'Relación entre variables climáticas y casos de dengue (lag 0)',
    fontsize=13, fontweight='bold'
)
plt.tight_layout()
plt.savefig('../reports/figures/07_scatter_clima_dengue.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 9. Matriz de correlaciones completa

Calculamos la correlación entre **todas** las variables del dataset de una vez.

La representamos como un heatmap triangular (mostramos solo el triángulo inferior para evitar duplicados).

**¿Para qué sirve esto en el modelado?**
- Identifica qué variables climáticas correlacionan más con los casos
- Detecta multicolinealidad (variables muy correlacionadas entre sí que pueden causar problemas en algunos modelos)
- Por ejemplo, temperatura media y máxima suelen tener r > 0.95 — incluir ambas puede ser redundante

```python
# Seleccionamos las columnas relevantes para la matriz
cols_corr = [
    'confirmed_cases', 'temp_mean', 'temp_max_mean', 'temp_min_mean',
    'precipitation', 'humidity_mean', 'heat_index_mean',
    'temp_mean_anomaly', 'precipitation_anomaly'
]

# .corr() calcula la correlación de Pearson entre todos los pares de columnas
corr_matrix = df[cols_corr].corr()

# Renombramos para que sea más legible en el gráfico
labels = {
    'confirmed_cases': 'Casos dengue',
    'temp_mean': 'Temp. media',
    'temp_max_mean': 'Temp. máx.',
    'temp_min_mean': 'Temp. mín.',
    'precipitation': 'Precipitación',
    'humidity_mean': 'Humedad',
    'heat_index_mean': 'Índice calor',
    'temp_mean_anomaly': 'Anomalía temp.',
    'precipitation_anomaly': 'Anomalía lluvia'
}
corr_matrix = corr_matrix.rename(index=labels, columns=labels)

fig, ax = plt.subplots(figsize=(11, 9))

# np.triu genera una máscara triangular superior (True = ocultar)
# Ocultamos el triángulo superior para no repetir información
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix,
    ax=ax,
    mask=mask,
    annot=True, fmt='.2f',     # mostrar el valor numérico en cada celda
    cmap='RdBu_r',              # rojo = correlación positiva, azul = negativa
    vmin=-1, vmax=1,
    center=0,                   # blanco en r=0 (sin correlación)
    square=True,
    linewidths=0.5,
    cbar_kws={'label': 'Correlación de Pearson (r)'}
)

ax.set_title('Matriz de correlaciones — dataset maestro', pad=15)
plt.tight_layout()
plt.savefig('../reports/figures/08_matriz_correlaciones.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 10. Incidencia relativa por comuna y año

La **incidencia relativa** normaliza los casos de cada comuna respecto al promedio de CABA.

- Valor = 1.0: la comuna tuvo exactamente el promedio de CABA
- Valor > 1.0: la comuna tuvo MÁS casos que el promedio (mayor riesgo)
- Valor < 1.0: la comuna tuvo MENOS casos que el promedio (menor riesgo)

**¿Por qué normalizar así?**
Porque si el 2024 tuvo el doble de casos que el 2023, una comuna que mantuvo proporciones similares tendrá incidencia relativa parecida en ambos años. Esto nos permite ver qué comunas son **estructuralmente** más vulnerables, independientemente del tamaño del brote.

```python
# Total de casos por comuna y año
df_incidencia = (
    df.groupby(['comuna_id', 'year'])['confirmed_cases']
    .sum()
    .reset_index()
)

# Para cada año, calculamos cuánto representa cada comuna respecto al promedio
# incidencia_relativa = casos_comuna / promedio_entre_comunas
# Si el promedio entre las 15 comunas en 2024 es 1585 casos,
# y la comuna 1 tuvo 3000, su incidencia relativa = 3000/1585 = 1.89
for year in df_incidencia['year'].unique():
    mask = df_incidencia['year'] == year
    media = df_incidencia.loc[mask, 'confirmed_cases'].mean()
    df_incidencia.loc[mask, 'incidencia_relativa'] = (
        df_incidencia.loc[mask, 'confirmed_cases'] / media
    )

fig, ax = plt.subplots(figsize=(14, 6))

# Transformamos a formato ancho para el heatmap
pivot_inc = df_incidencia.pivot(
    index='comuna_id', columns='year', values='incidencia_relativa'
)

sns.heatmap(
    pivot_inc,
    ax=ax,
    annot=True, fmt='.2f',
    cmap='RdYlGn_r',   # rojo = alta incidencia, verde = baja incidencia
    center=1,           # blanco en 1.0 (exactamente el promedio)
    vmin=0, vmax=3,
    linewidths=0.5,
    cbar_kws={'label': 'Incidencia relativa (1.0 = promedio CABA)'}
)

ax.set_xlabel('Año')
ax.set_ylabel('Comuna')
ax.set_title(
    'Incidencia relativa por comuna y año\n'
    '(rojo > promedio CABA, verde < promedio CABA)',
    pad=10
)

plt.tight_layout()
plt.savefig('../reports/figures/09_incidencia_relativa_comunas.png', dpi=150, bbox_inches='tight')
plt.show()

print('Comunas con incidencia relativa > 1.5 en 2024 (mayor riesgo estructural):')
año_2024 = df_incidencia[
    (df_incidencia['year'] == 2024) &
    (df_incidencia['incidencia_relativa'] > 1.5)
].sort_values('incidencia_relativa', ascending=False)
print(año_2024[['comuna_id', 'confirmed_cases', 'incidencia_relativa']].to_string(index=False))
```
